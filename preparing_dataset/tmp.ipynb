```python
import pandas as pd
from PIL import Image
import cv2
from PIL import ImageDraw
```

```python
def convert(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)

img=Image.open('/usr/src/datasets/rawVkitti/vkitti2.0.3rgb/Scene01/15-deg-left/frames/rgb/Camera_0/rgb_00000.jpg')
w= int(img.size[0])
h= int(img.size[1])
```

```python
b = (552,	590,	190,	212	)
bb = convert((w,h), b)

print(bb)
```

```python
draw = ImageDraw.Draw(img)
draw.rectangle([(b[0], b[2]), (b[1], b[3])], outline=(0, 255, 0))
img.show()
```
